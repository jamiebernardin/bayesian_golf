```python
import psycopg2 as pq
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
golf_db = 'postgresql://localhost:5432/golf'
pd.set_option('max_rows', 1000)
pd.set_option('max_columns', 50)
```

```python
def exec_sql(sqlTxt):  
    with pq.connect(golf_db) as conn:
        with conn.cursor() as cursor:
            cursor.execute(sqlTxt)    
            return [row for row in cursor]
def pd_sql(sqlTxt):
    with pq.connect(golf_db) as conn:
        return pd.read_sql_query(sqlTxt, conn)
    
def combine_rounds(rounds):
    def round_equals(row1, row2): 
        return row1[2] == row2[2] and row1[3] == row2[3] and row1[4] == row2[4]
    result = rounds[0]
    for round in rounds[1:]:
        head = result if np.shape(result) == np.shape(rounds[0]) else result[len (result) -1]
        if round_equals(head, round):
            head[5] += round[5]
        else:
            result = np.vstack((result, round))
    return result
   
```

```python
sqlTxt = '''
select    year, date, player_id, permanent_tournament_id, round,  count(shot) as score
  from    strokes 
    where year = 2016
      and player_id = 32102
    group by year, date, player_id, permanent_tournament_id, round
    order by date, permanent_tournament_id, round
'''
cursor = exec_sql(sqlTxt)
rounds = np.empty((1, 6))
for row in cursor:
    rounds = np.vstack((rounds, np.asarray(row).reshape((1,6))))
rounds = combine_rounds(rounds[1:,0:])
```

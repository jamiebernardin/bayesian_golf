```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
m = 100
dx = 1/(1.*m)
x = np.linspace(0, 1, m)
prior = np.exp(-((x-0.6)/0.3)**2)
likelihood = 5*stats.binom.pmf(6,9,x)
u_post = likelihood*prior
post = u_post/np.sum(u_post)/dx
plt.plot(x, prior)
plt.plot(x, likelihood)
plt.plot(x, post)
```

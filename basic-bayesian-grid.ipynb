```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
mu_actual, sigma_actual = -1, 4
n = 20
x = np.random.normal(mu_actual, sigma_actual, n)
y = np.linspace(1, 1, n)
plt.scatter(x, y)
```

```python
points = 100
sig_grid = np.linspace(0.1, 10.1, points)
mu_grid = np.linspace(-5, 5, points)
p = np.ndarray((points,points))
```

```python
def posterier(mu, sigma):
    likelihoods = stats.norm.pdf(x, mu, sigma)
    priors = 1 #10*stats.halfnorm.pdf(sigma, sigma_actual*3)
    return np.prod(likelihoods*priors)
```

```python
for i in range(0,points):
    for j in range(0,points):
        p[i, j] = posterier(mu_grid[i], sig_grid[j])
#         print mu_grid[i], sig_grid[j], p[i, j]
plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
plt.title('mu')
plt.plot(mu_grid, np.sum(p, axis=1))
plt.subplot(1, 2, 2)
plt.title('sigma')
plt.plot(sig_grid, np.sum(p, axis=0))
```
